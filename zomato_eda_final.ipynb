### 13. Using the box plot for good view of the data

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.boxplot(y=df['AverageCost'])
plt.title('Average Cost Box column')
plt.ylabel('Values')
plt.show()
```

### See the box plot and large number of outliers with the mean and min values spread in the data above

### 14, Lets use the histogram for the average cost column

```python
import  seaborn as sns
sns.histplot(sorted_new_df['AverageCost'], bins=100, color="black")
plt.title("Average Cost column histogram")

plt.show()
```

### We get to know that the values of 1000 are the most occured using the hist plot

```python
sns.scatterplot(x='Dinner Reviews', y='AverageCost', data=df)
plt.title("Scatter plot for Delivery Reviews and average cost")
plt.show()
```
